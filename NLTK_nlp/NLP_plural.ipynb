```python
import urllib2
from bs4 import BeautifulSoup
```

```python
url="https://www.washingtonpost.com/politics/trumps-arrival-on-eve-of-inauguration-ushers-in-real-change-to-washington/2017/01/19/f2134a40-de62-11e6-918c-99ede3c8cafa_story.html?hpid=hp_hp-banner-main_inauguralweb-banner-515p%3Ahomepage%2Fstory&utm_term=.5414e65a447f"
```

```python
page=urllib2.urlopen(url).read().decode('utf','ignored')
```

```python
soup=BeautifulSoup(page)

```

```python
for texts in soup.find_all('article'):
    texts=texts.text
text=''.join(texts)
print text
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
from nltk.corpus import stopwords
from string import punctuation
text_encode=text.encode("ascii",errors="replace").replace("?"," ")
```

```python
sents=sent_tokenize(text_encode)
```

```python
sents
```

```python
words_token=word_tokenize(text_encode.lower())
words_token
```

```python
stopwords_panct=set(stopwords.words('english')+list(punctuation))
```

```python
word_sent=[word for word in words_token if word not in stopwords_panct]
word_sent
```

```python
from nltk.probability import FreqDist
freq=FreqDist(word_sent)
freq
```

```python
freq.most_common(20)
```

```python
import heapq
from collections import defaultdict
```

```python
heapq.nlargest(10,freq,key=freq.get)
```

```python
ranking=defaultdict(int)
```

```python
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i]+=freq[w]
sent_idx=heapq.nlargest(4,ranking,key=ranking.get)
print sent_idx
print [sents[j] for j in sorted(sent_idx)]
```
